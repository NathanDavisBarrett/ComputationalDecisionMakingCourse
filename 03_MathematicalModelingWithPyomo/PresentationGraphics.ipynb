```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,1,figsize=(3,9))

ax1.plot([0,1,2],[0.5,0,1])
ax1.plot([0,0.75,2],[-0.25,0.5,0.25])
ax1.fill_between([0,1,2],[-0.25,-0.25,-0.25],[0.5,0,1],color="grey",alpha=1)
ax1.fill_between([0,0.75,2],[1,1,1],[-0.25,0.5,0.25],color="grey",alpha=1)
ax1.fill_between([-10.-10],[-10,-10],[-10,-10],color="white",edgecolor="black",label="Feasible Region")

arrow = mpatches.FancyArrowPatch((0.6,0.25),(1,0.3),mutation_scale=10,color="green",label="Objective")
ax1.add_patch(arrow)

ax1.set_xlim(0,2)
ax1.set_ylim(-0.25,1)
ax1.axis(False)
ax1.legend(loc='upper left')
ax1.set_title("Feasible Problem")

ax2.plot([0,1,2],[0.75,0.25,1.25])
ax2.plot([0,0.75,2],[-0.5,0.25,0.])
ax2.fill_between([0,1,2],[-0.25,-0.25,-0.25],[0.75,0.25,1.25],color="grey",alpha=1)
ax2.fill_between([0,0.75,2],[1,1,1],[-0.5,0.25,0.],color="grey",alpha=1)
ax2.fill_between([-10.-10],[-10,-10],[-10,-10],color="white",edgecolor="black",label="Feasible Region")

arrow = mpatches.FancyArrowPatch((0.4,0.25),(0.8,0.3),mutation_scale=10,color="green",label="Objective")
ax2.add_patch(arrow)

ax2.set_xlim(0,2)
ax2.set_ylim(-0.25,1)
ax2.axis(False)
ax2.legend(loc='upper left')
ax2.set_title("Infeasible Problem")

ax3.plot([0,1,2],[0.5,0,0.15])
ax3.plot([0,0.75,2],[-0.25,0.5,0.85])
ax3.fill_between([0,1,2],[-0.25,-0.25,-0.25],[0.5,0,0.15],color="grey",alpha=1)
ax3.fill_between([0,0.75,2],[1,1,1],[-0.25,0.5,0.85],color="grey",alpha=1)
ax3.fill_between([-10.-10],[-10,-10],[-10,-10],color="white",edgecolor="black",label="Feasible Region")

arrow = mpatches.FancyArrowPatch((0.6,0.25),(1,0.3),mutation_scale=10,color="green",label="Objective")
ax3.add_patch(arrow)

ax3.set_xlim(0,2)
ax3.set_ylim(-0.25,1)
ax3.axis(False)
ax3.legend(loc='upper left')
ax3.set_title("Unbounded Problem")

fig.savefig("InfeasibleUnbounded.png")
```
